```python
08.05 스튜던트 t분포, 카이제곱분포, F분포


0) 모수(parameter), 통계량(statistics) 비교 (2page)








- 정규분포에서 파생된 분포 : 정규분포에서 생성된 데이터에 수식적용하면 분포모양이 달라짐
    - 적용된 수식에 따라 스튜던트t, 카이제곱, F분포
    - 즉, 다 ' 통계량 분포' 이다!
    
    
1) 스튜던트 t분포 (정규분포에서 얻은 표본평균을 표본표준편차로 정규화   한 통계량이 확률변수)
    - 분포가 fat-tail (꼬리가 두꺼운) 모형에 적합한 것이 스튜던트 t 분포(t분포)
    - t분포의 pdf (모수 : 기댓값, 표준편차, 자유도)
    - 자유도 1 = 코시분포 (양수만 다루면 하프코시분포)
    - 자유도에 따른 스튜던트 t 분포의 변화 (자유도가 커질수록 fat-tail사라지며 정규분포화!) (4paage)
    
    
    
    
    
    
    - t분포의 모멘트 (8.5.3 ~ 4)
    
    
    
    
    
2) t통계량
    - z 통계량을 구하려면 확률분포(X, 이상)의 정확한 평균, 표준편차를 알아야 한다. 이는 어려움
    - 따라서, 표본평균을 표본표준편차로 정규화한 값 = t통계량 (z통계량의 표본버젼) (8.5.5 ~ 7)
       
    
    
    
    
    - 하지만, 자유도를 30이상으로 높이면, t분포가 정규분포와 거의 동일 <=> t통계량 = z통계량 (6page)
    
    
    
    
    
    
    
3) 카이제곱 분포 (정규분포에서 얻은 변수를 제곱+SUM   한 통계량이 확률변수)
    - 8.5.8 ~ 10
    
    
    
    
    - 자유도에 따른 카이제곱 분포의 변화 (8 page)
    (많이 뽑을 수록 카이제곱 확률변수(통계량)의 분포는 정규분포화 된다.)
    
    
    
    
    
    
    
    
    
    
4) F 분포 (카이제곱분포를 따르는 2개의 확률변수를 각각 N(표본갯수)으로 나눠 비교   한 통계량이 확률변수)
    - t분포, 카이제곱분포 = 정규분포 확률변수에서 N개의 표본을 추출해 생성
    - F 분포 = 카이제곱분포에서 확률변수 2개 뽑아서 진행. 즉, N1+N2 개의 표본을 추출해 생성
    - 즉, F 분포 = 제곱합 비율의 분포
    - 8.5.12
    
    
    
    - 11page
    
    
    
    
    
    
    
    
    
    
    
5) 통계량 분포 정리 (암기) 13page
    

    
    
    
    
    
    
    
    
    
    
```
