# 1. Data Cleaning _ Price

1) 자료파악
    - 0과 그외 (무료앱 = 0)
    - unique로 특성 파악 ($, everyone 등 이상치 파악)
    
2) 이상치 제거 ($, everyone)
3) 특이행 확인 (100불이상) -> 제거하지는 않음. 특성만 파악함

4) 자료형 변환 (object -> float64)

5) 결측치 확인
    - 없음
    - 유료앱 : 800개, 그외 무료앱

6) 정리

    - 1) 전체 행 : 10840행
    - 2) 무료 / 유료 : 10040, 800
    - 3) 최대값 : 400불
    - 4) 유료 앱의 평균가격 : 약 14불
    - 5) 결측치 없음
    - 6) 제거 행 : 10472행 (컬럼이 밀려서 사용할 수 없는 데이터)

```python
# 자료 파악

data.Price.unique()
```

```python
# $표시 제거 -> apply 함수 활용

data.Price = data.Price.apply(lambda x: x.replace('$',''))
data.Price.unique()
```

```python
# Everyone -> 행 자체를 제거. 컬럼이 1개씩 밀려서 사용할 수 없는 행. 10472행 제거

data[data.Price == 'Everyone']
```

```python
# Everyone -> 행 자체를 제거. 컬럼이 1개씩 밀려서 사용할 수 없는 행. 10472행 제거 -> data_1 생성

data_1 = data.drop(10472)
data_1.shape
```

```python
data_1.Price.unique()
```

```python
# 자료형 변환

data_1.Price.dtypes
```

```python
# 자료형 변환

data_1.Price = pd.to_numeric(data.Price, errors='coerce')
data_1.Price.dtypes
```

```python
# 이상치 확인 -> 100불 이상

temp = data_1.Price.apply(lambda x: True if x>100 else False)
data_1[temp]
```
