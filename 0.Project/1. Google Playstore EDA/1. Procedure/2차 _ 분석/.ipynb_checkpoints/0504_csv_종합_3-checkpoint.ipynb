```python
data = pd.read_csv('/Users/kimjeongseob/Desktop/Github upload/5.EDA/EDA 프로젝트/2차 _ 분석/csv 모음/데이터_2_new.csv')
data.drop(columns = ['Unnamed: 0'])
```

```python
data.info()
```

```python
data['Content Rating'] = data['Content Rating'].astype('string')
data['Genres'] = data['Genres'].astype('string')
data['pri_genres'] = data['pri_genres'].astype('string')
data['sec_genres'] = data['sec_genres'].astype('string')
data.info()
```

   # 1. Install 편차 줄이기

```python
# install 숫자가 너무 편차가 크다...
# 백만으로 나누던가 해서 그 숫자 자체를 줄여줄 필요가 있다.
# 일단 install 의 분포가 어떤지 보자

# 1. install의 분포
data['Installs'].value_counts()
```

```python
# 백만개 이상 인스톨 = 2832 행
# 만개 이상 인스톨 = 6598 행

data['Installs_mil'] = data['Installs']/1000000
data['Installs_mil'].value_counts()
# data[data['Installs_mil'] > 1]

data['Installs_10t'] = data['Installs']/10000
# data[data['Installs_10t'] > 1]
data['Installs_10t'].value_counts()

data['Installs_bil'] = data['Installs']/1000000000
```

```python
# 전체적인 상관관계 분석 - seaborn, heatmap
# 참고 : https://www.kaggle.com/mhmdpkts/playstore-preprocessing-and-eda


fit, ax = plt.subplots(figsize = (10,8))
ax = sns.heatmap(data[['Installs_10t','Rating','reviews','Size','Price']].corr(), annot=True, Linewidths=.5, fmt='.1f')
plt.show()

# 인스톨 - 리뷰 간 계수 0.6, 나머진 거의 없음..
```

# 카테고리 - count

```python
# 카테고리별 앱 갯수

plt.figure(figsize = (10,10))
g = sns.countplot(x='Category',data = data)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
plt.title('Count of app in each category', size=20)
```

```python
# category 별 rate
# rate... 하위 25% 모두 0인 것들이 있는데.. 전처리할 때 value 조정한 것 있었는 지?

g = sns.catplot(x="Category",y="Rating",data=data, kind="box", height = 10 ,
palette = "Set1")
g.despine(left=True)
g.set_xticklabels(rotation=90)
g.set( xticks=range(0,34))
g = g.set_ylabels("Rating")
plt.title('Boxplot of Rating VS Category',size = 20)
```
