```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import missingno as msno
```

```python
a1=pd.read_csv('googleplaystore.csv')
a1
```

```python

caa={'Rating':0,'Type':'Free','Content Rating':'Everyone','Current Ver':'Varies with device','Android Ver':'Varies with device','Size':'37M'}
ad=a1.fillna(value=caa)
ad
```

```python
len(ad[ad["Size"].isnull()])
```

```python
msno.bar(ad)
```

```python
ad
```

```python
ad.drop(['Last Updated','Current Ver','Android Ver'],axis=1,inplace=True)
```

```python
ad
```

```python
ad.columns = ad.columns.str.replace(" ","_")
```

```python
ad.Type=ad.Type.replace("Free",0)
ad.Type=ad.Type.replace("Paid",1)

ad.Size = ad.Size.replace("Varies with device",np.nan)
ad.Size = ad.Size.str.replace("M","")
ad.Size = ad.Size.str.replace("k","")


#store.Size = store.Size.apply(lambda x: float(x.replace("k",""))/1000 if "k" in x else x)
#ı wanted use this but ı get error so ı added "000" to Megabyte

ad.Size = ad.Size.replace("1,000+",1000)

ad.Installs = ad.Installs.str.replace(",","")
ad.Installs = ad.Installs.apply(lambda x: x.strip("+"))
ad.Installs = ad.Installs.replace("Free",np.nan)

ad.Reviews = ad.Reviews.replace("Varies with device",np.nan)
ad.Reviews = ad.Reviews.str.replace("M","")

ad.Price = ad.Price.str.replace("$","")

ad = ad.drop(ad.index[10472])
ad= ad.drop(ad.index[9148])  #삭제행 정하기 , 


ad[["Size","Installs","Reviews","Price","Type"]] = ad[["Size","Installs","Reviews","Price","Type"]].astype("float")
```

```python
# null 확인 
mask_t2 = pd.to_numeric(ad["Price"], errors='coerce').isna()
print(mask_t2.sum())

mask_t = pd.to_numeric(ad["Size"], errors='coerce').isna()
print(mask_t.sum())
```

```python
# 사이즈 null 값 대채 
average_size = round(np.average(ad.Size.dropna()), 1)
average_size
```

```python
ad["Size"][ad["Size"].isnull()] = average_size
```
