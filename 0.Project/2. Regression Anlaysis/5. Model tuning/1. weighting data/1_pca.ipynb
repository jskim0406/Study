```python
df_copy = df_copy.reset_index()
```

```python
df_copy_info = df_copy[['player_name','value','follower','age','rating']]
df_copy_atk = df_copy[['goals_total', 'goals_assists']]
df_copy_def = df_copy[['tackles_blocks','tackles_interceptions']]
df_copy_contribution = df_copy[['duels_won','dribbles_success','fouls_drawn', 'shots_on', 'passes_key']]
df_copy_mistake = df_copy[['fouls_committed', 'cards_yellow', 'cards_yellowred', 'cards_red']]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1) # 주성분을 몇개로 할지 결정
# 주성분으로 이루어진 데이터 프레임 구성
printcipalComponents = pca.fit_transform(df_copy_atk)
principalDf = pd.DataFrame(data=printcipalComponents, columns = ['attacks'])

printcipalComponents1 = pca.fit_transform(df_copy_def)
principalDf1 = pd.DataFrame(data=printcipalComponents1, columns = ['defenses'])

printcipalComponents2 = pca.fit_transform(df_copy_contribution)
principalDf2 = pd.DataFrame(data=printcipalComponents2, columns = ['contributions'])

printcipalComponents3 = pca.fit_transform(df_copy_mistake)
principalDf3 = pd.DataFrame(data=printcipalComponents3, columns = ['mistakes'])
```

```python
merged_1 = pd.concat([principalDf, principalDf1], axis=1)
merged_2 = pd.concat([merged_1, principalDf2], axis=1)
merged_3 = pd.concat([merged_2, principalDf3], axis=1)
pca_data = pd.concat([df_copy_info, merged_3], axis=1).set_index('player_name')

pca_data
```

```python
pca_data.follower = np.log(pca_data.follower)
pca_data.follower = pca_data.follower.replace(-np.inf,0)
```

```python
pca_data.corr()[pca_data.corr() > 0.5]
```

```python
pca_data_x = pca_data.drop('value', axis=1)
pca_data_y = pca_data.value
```

```python
N = len(pca_data)
ratio = 0.7
np.random.seed(0)
idx_train = np.random.choice(np.arange(N), np.int(ratio * N))
idx_test = list(set(np.arange(N)).difference(idx_train))

df_train = pca_data.iloc[idx_train]
df_test = pca_data.iloc[idx_test]
```

```python
import statsmodels.api as sm

model_pca = sm.OLS.from_formula('value ~ attacks+np.log(attacks)+defenses+mistakes+np.log(mistakes)+\
contributions+age+rating+follower',data = df_train).fit()
print(model_pca.summary())
```

```python
df_test = pd.DataFrame(df_test)

pred = model_pca.predict(df_test)

rss = ((df_test.value - pred) ** 2).sum()
tss = ((df_test.value - df_test.value.mean()) ** 2).sum()
rsquared = 1 - rss / tss
rsquared
```

```python
from sklearn.model_selection import KFold

scores = np.zeros(5)
cv = KFold(5, shuffle=True, random_state=0)
for i, (idx_train, idx_test) in enumerate(cv.split(pca_data)):
    df_train = pca_data.iloc[idx_train]
    df_test = pca_data.iloc[idx_test]
       
    pred = model_pca.predict(df_test)
    rss = ((df_test.value - pred) ** 2).sum()
    tss = ((df_test.value - df_test.value.mean()) ** 2).sum()
    rsquared = 1 - rss / tss
    
    scores[i] = rsquared
    print("학습 R2 = {:.8f}, 검증 R2 = {:.8f}".format(model_pca.rsquared, rsquared))
```

```python
import statsmodels.api as sm

model = sm.OLS.from_formula('value ~ attacks+defenses+mistakes+\
contributions+age+rating+follower',data = df_train)
result = model.fit()
print(result.summary())
```

```python
df_test = pd.DataFrame(df_test)

pred = result.predict(df_test)

rss = ((df_test.value - pred) ** 2).sum()
tss = ((df_test.value - df_test.value.mean()) ** 2).sum()
rsquared = 1 - rss / tss
rsquared
```
