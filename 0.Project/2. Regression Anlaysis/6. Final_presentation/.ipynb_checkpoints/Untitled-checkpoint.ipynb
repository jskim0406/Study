```python
folx_ols_1 = 0.14
folo_ols_1 = 0.34
folx_ols_2 = 0.13
folx_reg_2 = 0.0367
folo_ols_2 = 0.25
folo_reg_2 = 0.202
folx_ols_3 = 0.2
folx_reg_3 = 0.276
folo_ols_3 = 0.32
folo_reg_3 = 0.365
folx_ols_4 = 0.085
folo_ols_4 = 0.26

folx = list([folx_ols_1,folx_ols_2,folx_reg_2,folx_ols_3,folx_reg_3,folx_ols_4])
folo = list([folo_ols_1,folo_ols_2,folo_reg_2,folo_ols_3,folo_reg_3,folo_ols_4])
```

```python
import plotly.graph_objects as go
models=['1. Domain', '2. Domain(expanded)','2. Ridge','3. PCA','3. Ridge','4. PCA(attacker)']

fig = go.Figure(data=[go.Bar(name='follower(X)', x=models, y=folx),go.Bar(name='follower(O)', x=models, y=folo)])
# Change the bar mode
fig.update_layout(barmode='group',xaxis_tickangle=-20)
fig.update_layout(title_text='R2_score of Models(OLS, Regularized Regression)')
fig.show()
```
