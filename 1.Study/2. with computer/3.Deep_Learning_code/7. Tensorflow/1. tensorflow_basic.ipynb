### 2. Eager mode

- 즉시 실행 모드

````python
a = tf.constant(3)
b = tf.constant(2)
````

````python
# 기본 연산
# tensor로 출력
print(tf.add(a,b))
print(tf.subtract(a,b))
````

````python
# numpy로 출력
# 보통은, numpy 변환 후(.numpy()) 함수 연산 -> 다시 텐서로 변환(tf.convert_to_tensor())

print(tf.multiply(a,b).numpy())
print(tf.divide(a,b).numpy())
````

````python
c = tf.add(a,b).numpy()
c_square = np.square(c, dtype=np.float32)
c_tensor = tf.convert_to_tensor(c_square)
````

````python
print("numpy_array : %0.1f, applying square with numpy : %0.1f,\
      convert_to_tensor : %0.1f" % (c, c_square, c_tensor))
````

````python
tf.rank(c_square)
````

### 3. @tf.function

- **graph형태로 다루고 싶을 때 사용**
- Auto graph : 자동으로 그래프를 생성해주는 기능(데코레이터)
- 이를 통해, 파이썬 코드를 그래프로 변환해 사용 가능 -> 그래프로 변환해 사용한다는 것은 gpu연산이 가능하다는 의미(효율적)
- 하지만, CNN 등의 고비용 연산에서는 성능에 큰 차이를 갖지 못한다는 단점

    - if : tf.cond
    - for/while : tf.while_loop
    - for _ in dataset : dataset.reduce

````python
@tf.function
def square_pos(x):
    if x>0:
        x = x*x
    else:
        x = x*(-1)
    return x

square_pos
````

### 4. OR 게이트 구현

````python
tf.random.set_seed(777)
````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse
````

````python
# 데이터셋 준비
x = np.array([[0,0],[1,0],[0,1],[1,1]])
y = np.array([[0],[1],[1],[1]])
````

````python
# 모델 구성
model = Sequential()

# Dense(hidden_unit 갯수, input_shape) : 1개의 hidden_unit, 2개의 특성을 갖는 1차원 데이터(1-d tensor)를 입력으로 받고, 1개의 출력을 갖는 dense 층
# hidden_unit 갯수 : output size에 영향
# ex) input : [1,0], output : 1
model.add(Dense(1,input_shape=(2,),activation='linear'))
````

````python
# 모델 준비하기
model.compile(optimizer=SGD(),
              loss = mse,
              metrics = ['acc']) # acc metric은 list 형태로 평가지표를 전달함
````

````python
# 학습 시키기
model.fit(x,y,epochs=500)
````

````python
# 가중치 확인
model.get_weights()
````

### 6. XOR + MLP

````python
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.losses import mse

tf.random.set_seed(777)

# 데이터 준비
data = np.array([[0,0],[1,0],[0,1],[1,1]])
label = np.array([[0],[1],[1],[0]])

# 모델 구성
model = Sequential()
model.add(Dense(32, input_shape=(2,), activation = 'relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델 준비하기
model.compile(optimizer=RMSprop(),
              loss = mse,
              metrics = ['acc'])

# 모델 학습
model.fit(data,label, epochs=1000)
````

### 8. practice

- Perceptron_single-layer (XOR problem)

````python
data = np.array([[0,0],[1,0],[0,1],[1,1]])
label = np.array([[0],[1],[1],[0]])
````

````python
# 모델 구성
model_single = Sequential()
model_single.add(Dense(1, input_shape=(2,), activation='linear'))

model_MLP = Sequential()
model_MLP.add(Dense(32, input_shape=(2,), activation='relu'))
model_MLP.add(Dense(1, activation='sigmoid'))

# 학습 설정
model_single.compile(optimizer='sgd', loss='mse', metrics=['acc'])

model_MLP.compile(optimizer=RMSprop(), loss=mse, metrics=['acc'])

# 학습
model_single.fit(data, label, epochs=100)
````

````python
# 평가
model_single.evaluate(data, label)
````

````python
# 예측
model_single.predict(data)
````

````python
# MLP 학습, 예측
model_MLP.fit(data, label, epochs=100)
````

````python
model_MLP.predict(data)
````
