```
str.lower() : 소문자로 변환 (in-place X)
str.upper() : 대문자로 변환 (in-place X)

str.split() : 공백 기준 str 쪼개기
str.split('a') : 'a'를 기준으로 str 쪼개기 *('a'는 제거됨)
str.replace('a','b') : 'a'를 'b'로 변환 (in-place 됨)
str.find('a') : 'a'의 index 반환
str.rfind('a') : 'a'의 index 반환(뒤에서 부터 탐색)
str.capitalize() : 문장열 중에서 첫번째 단어만 대문자로 바꾼다

str.join() : 두 문자열의 결합  ex) " ".join("abc") => a b c
str.join([str1, str2]) : 문자열을 요소로 갖는 리스트를 새로운 결합된 문자열로 변환   ex) ", ".join(['a','b']) => a, b
"".join(reversed(str)) : 문자열을 뒤집어 반환  ex) "".join(reversed("abc")) => cba
str.title() : 문자 열중 각각의 단어의 첫 부분만 대문자로 바꾼다.
str.swapcase() : 문자열중 소문자는 대문자로 , 대문자는 소문자로 바꾼다. 
str.count() : arg 가 문자열 중에서 몇 번 들어가는지 세어서 반환한다.
str.center(10) : 문자열을 차지하는 공간(10)에서 문자열들을 중앙에 배치한다.
str.ljust(10) : 문자열을 차지하는 공간(10)에서 문자열을 좌측으로 정렬한다.
str.rjust(10) : 문자열을 차지하는 공간(10)에서 문자열을 우측으로 정렬한다.


len(str) : 문자열 길이 반환

[boolean]
    str.isalnum()	문자열은 영문자와 숫자로만 구성되어 있는가? (no symbols)
    str.isalpha()	문자열이 영문자로만 이루어져 있는가? (no symbols)
    str.islower()	문자열에 포함된 영문자가 모두 소문자인가?
    str.isnumeric()	문자열이 숫자로만 구성되어 있는가?
    str.isspace()	문자열이 공백문자(white space)로만 구성되어 있는가?
    str.istitle()	문자열이 제목 형식(단어의 첫문자가 대문자)에 있는가?
    str.isupper()	문자열에 포함된 영문자가 모두 대문자인가?
    str.startswith('arg') : arg(인자)가 문자열 내에서 처음으로 나오면 True 아니면 False 
    str.endswith('arg') : arg(인자) 가 문자열 내에서 마지막으로 나오면 True 아니면 False
```
