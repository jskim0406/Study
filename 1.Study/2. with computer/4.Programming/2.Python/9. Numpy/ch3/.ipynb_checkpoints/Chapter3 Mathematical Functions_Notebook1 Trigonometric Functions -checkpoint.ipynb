# Trigonometric functions 삼각함수

```
input : radian

1. np.sin()
2. np.cos()
3. np.tan()
4. np.arcsin()
5. np.arccos()
6. np.arctan()
7. np.deg2rad()
8. np.rad2deg()
```

````python
import math
PI = math.pi
PI
````

````python
radian_range = np.linspace(0, PI*4, 400)
````

### 1. np.sin / np.cos / np.tan

````python
sin = np.sin(radian_range)
cos = np.cos(radian_range)
tan = np.tan(radian_range)
````

````python
%matplotlib inline
plt.figure(figsize=(15,10))
plt.ylim(-2,2)
plt.plot(sin, label='sin')
plt.plot(cos, label='cos')
plt.plot(tan, label='tan')
plt.legend(fontsize=20)
plt.show()
````

### 2. np.arcsin / np.arccos / np.arctan

- 역 삼각함수

````python
radian_range = np.linspace(-1,1,300)
````

````python
arcsin = np.arcsin(radian_range)
arccos = np.arccos(radian_range)
arctan = np.arctan(radian_range)
````

````python
%matplotlib inline
plt.figure(figsize=(15,10))
plt.ylim(-2,4)
plt.plot(arcsin, label='arcsin')
plt.plot(arccos, label='arccos')
plt.plot(arctan, label='arctan')
plt.legend(fontsize=20)
plt.show()
````

### 3. degree <-> radian

- 각도 표기를 degree와 radian표기로 서로 switching

````python
deg = [0,90,180,270,360]
rad = np.deg2rad(deg)
rad
````
