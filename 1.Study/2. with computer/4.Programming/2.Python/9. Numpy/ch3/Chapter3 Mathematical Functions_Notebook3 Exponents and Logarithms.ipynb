# Exponential, Log


1. np.exp() : $e^{x}$
2. np.exp2() : $2^{x}$
3. np.log() : $logx$
4. np.log2() : $log_{2}x$
5. np.log10() : $log_{10}x$

```python
np.exp(1)
```

```python
np.exp2(1)
```

```python
%matplotlib inline
range_ = np.linspace(0,10,1000)

exp = np.exp(range_)
exp2 = np.exp2(range_)
log = np.log(range_)
log2 = np.log2(range_)
log5 = np.log(range_) / np.log(5)  # ë°‘ = 5

plt.figure(figsize=(15,10))
plt.ylim(-10,10)
plt.plot(exp, label='exp')
plt.plot(exp2, label='exp2')
plt.plot(log, label='log')
plt.plot(log2, label='log2')
plt.plot(log5, label='log5')
plt.legend()
```
