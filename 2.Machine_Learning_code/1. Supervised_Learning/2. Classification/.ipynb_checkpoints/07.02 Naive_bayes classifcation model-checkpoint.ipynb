```
[ÏÇ¨Ïù¥ÌÇ∑Îü∞ÏóêÏÑú Ï†úÍ≥µÌïòÎäî ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï]

Ï¢ÖÏÜçÎ≥ÄÏàò YÏùò Î¨¥Ï°∞Í±¥Î∂Ä ÌôïÎ•†Î∂ÑÌè¨  ùëÉ(ùëå)  (Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò Í≤ΩÏö∞ÏóêÎßå)


class_log_prior_
Ï¢ÖÏÜçÎ≥ÄÏàò YÏùò Î¨¥Ï°∞Í±¥Î∂Ä ÌôïÎ•†Î∂ÑÌè¨Ïùò Î°úÍ∑∏  logùëÉ(ùëå)  (Î≤†Î•¥ÎàÑÏù¥Î∂ÑÌè¨ÎÇò Îã§Ìï≠Î∂ÑÌè¨Ïùò Í≤ΩÏö∞ÏóêÎßå)




[Ï†ïÍ∑úÎ∂ÑÌè¨ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï]

Í∞ÄÏö∞ÏãúÏïà ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï GaussianNBÏùÄ Í∞ÄÎä•ÎèÑ Ï∂îÏ†ïÍ≥º Í¥ÄÎ†®ÌïòÏó¨ Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏÜçÏÑ±ÏùÑ Í∞ÄÏßÑÎã§.

theta_: Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò Í∏∞ÎåìÍ∞í  ùúá 
sigma_: Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò Î∂ÑÏÇ∞  ùúé2



[Î≤†Î•¥ÎàÑÏù¥ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï]

Î≤†Î•¥ÎàÑÏù¥Î∂ÑÌè¨ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï ÌÅ¥ÎûòÏä§ BernoulliNBÎäî Í∞ÄÎä•ÎèÑ Ï∂îÏ†ïÍ≥º Í¥ÄÎ†®ÌïòÏó¨ Îã§Ïùå ÏÜçÏÑ±ÏùÑ Í∞ÄÏßÑÎã§.

feature_count_: Í∞Å ÌÅ¥ÎûòÏä§  ùëò Ïóê ÎåÄÌï¥  ùëë Î≤àÏß∏ ÎèôÏ†ÑÏù¥ ÏïûÎ©¥Ïù¥ ÎÇòÏò® ÌöüÏàò  ùëÅùëë,ùëò 
feature_log_prob_: Î≤†Î•¥ÎàÑÏù¥Î∂ÑÌè¨ Î™®ÏàòÏùò Î°úÍ∑∏

```

# 1. Ï†ïÍ∑úÎ∂ÑÌè¨ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï


    - Îç∞Ïù¥ÌÑ∞ : Ïã§Ïàò, ÌÅ¥ÎûòÏä§ ÎßàÎã§ ÌäπÏ†ïÌïú Í∞í Ï£ºÎ©¥ÏóêÏÑú Î∞úÏÉùÌï† Í≤ΩÏö∞
    - Ï†ïÍ∑úÎ∂ÑÌè¨ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï : Ï†ïÍ∑úÎ∂ÑÌè¨ Í∞ÄÎä•ÎèÑÎ™®ÌòïÏùÑ Ï†ÅÏö©Ìïú ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Îç∏
        - Ï†ïÍ∑úÎ∂ÑÌè¨ Í∞ÄÎä•ÎèÑ Î™®Ìòï : Îã§Î≥ÄÏàò Ï†ïÍ∑úÎ∂ÑÌè¨ pdfÏóê Ï°∞Í±¥Î∂Ä ÎèÖÎ¶Ω Ï†ÅÏö©
        

**Î≥∏Îûò Îã§Î≥ÄÏàò Ï†ïÍ∑úÎ∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞**

````python
np.random.seed(0)
rv0 = sp.stats.multivariate_normal([-2, -2], [[1, 0.9], [0.9, 2]])
rv1 = sp.stats.multivariate_normal([2, 2], [[1.2, -0.8], [-0.8, 2]])
X0 = rv0.rvs(40)
X1 = rv1.rvs(60)
X = np.vstack([X0, X1])
y = np.hstack([np.zeros(40), np.ones(60)])

xx1 = np.linspace(-5, 5, 100)
xx2 = np.linspace(-5, 5, 100)
XX1, XX2 = np.meshgrid(xx1, xx2)
plt.grid(False)
plt.contour(XX1, XX2, rv0.pdf(np.dstack([XX1, XX2])), cmap=mpl.cm.cool)
plt.contour(XX1, XX2, rv1.pdf(np.dstack([XX1, XX2])), cmap=mpl.cm.hot)
plt.scatter(X0[:, 0], X0[:, 1], marker="o", c='b', label="y=0")
plt.scatter(X1[:, 0], X1[:, 1], marker="s", c='r', label="y=1")
plt.legend()
plt.title("Îç∞Ïù¥ÌÑ∞Ïùò ÌôïÎ•†Î∂ÑÌè¨")
plt.axis("equal")
plt.show()
````

**Î≥∏Îûò Îã§Î≥ÄÏàò Ï†ïÍ∑úÎ∂ÑÌè¨ -> ÎÇòÏù¥Î∏å Í∞ÄÏ†ïÏùÑ Ï†ÅÏö©Ìïú ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®ÌòïÏúºÎ°ú Ìï¥ÏÑù**

````python
from sklearn.naive_bayes import GaussianNB
model_norm = GaussianNB().fit(X, y)
````

````python
model_norm.classes_
````

````python
model_norm.class_count_
````

````python
model_norm.class_prior_
````

````python
model_norm.theta_[0], model_norm.sigma_[0]
````

````python
rv0 = sp.stats.multivariate_normal(model_norm.theta_[0], model_norm.sigma_[0])
rv1 = sp.stats.multivariate_normal(model_norm.theta_[1], model_norm.sigma_[1])

xx1 = np.linspace(-5, 5, 100)
xx2 = np.linspace(-5, 5, 100)
XX1, XX2 = np.meshgrid(xx1, xx2)
plt.grid(False)
plt.contour(XX1, XX2, rv0.pdf(np.dstack([XX1, XX2])), cmap=mpl.cm.cool)
plt.contour(XX1, XX2, rv1.pdf(np.dstack([XX1, XX2])), cmap=mpl.cm.hot)
plt.scatter(X0[:, 0], X0[:, 1], marker="o", c='b', label="y=0")
plt.scatter(X1[:, 0], X1[:, 1], marker="s", c='r', label="y=1")

x_new = [0, 0]
plt.scatter(x_new[0], x_new[1], c="g", marker="x", s=150, linewidth=5)
plt.legend()
plt.title("ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶àÎ°ú Ï∂îÏ†ïÌïú Îç∞Ïù¥ÌÑ∞Ïùò ÌôïÎ•†Î∂ÑÌè¨")
plt.axis("equal")
plt.show()
````

### Ïó∞ÏäµÎ¨∏Ï†ú

Î∂ìÍΩÉ Î∂ÑÎ•òÎ¨∏Ï†úÎ•º Í∞ÄÏö∞ÏãúÏïà ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®ÌòïÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÌíÄÏñ¥Î≥¥Ïûê.

(1) Í∞ÅÍ∞ÅÏùò Ï¢ÖÏù¥ ÏÑ†ÌÉùÎê† ÏÇ¨Ï†ÑÌôïÎ•†ÏùÑ Íµ¨ÌïòÎùº.

(2) Í∞ÅÍ∞ÅÏùò Ï¢ÖÏóê ÎåÄÌï¥ ÍΩÉÎ∞õÏπ®Ïùò Í∏∏Ïù¥, ÍΩÉÎ∞õÏπ®Ïùò Ìè≠, ÍΩÉÏûéÏùò Í∏∏Ïù¥, ÍΩÉÏûéÏùò Ìè≠Ïùò ÌèâÍ∑†Í≥º Î∂ÑÏÇ∞ÏùÑ Íµ¨ÌïòÎùº.

(3) ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©ÌïòÏó¨ Î∂ÑÎ•òÎ¨∏Ï†úÎ•º ÌíÄÍ≥† Îã§ÏùåÏùÑ Í≥ÑÏÇ∞ÌïòÎùº.

- Î∂ÑÎ•òÍ≤∞Í≥ºÌëú 
- Î∂ÑÎ•òÎ≥¥Í≥†ÏÑú


````python
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB

# 1. Îç∞Ïù¥ÌÑ∞ Î°úÎìú

iris = load_iris()
X = np.array(iris.data)
y = np.array(iris.target)

# 2. naive-bayes ÌïôÏäµ
model = GaussianNB().fit(X, y)

# 3. ÏÇ¨Ï†ÑÌôïÎ•† Íµ¨ÌïòÍ∏∞
model.class_prior_
````

````python
# 4. Í∞ÅÍ∞ÅÏùò Ï¢Ö(classÎ≥Ñ) featureÏùò ÌèâÍ∑†, Î∂ÑÏÇ∞(Î™®Ïàò)Î•º Íµ¨ÌïòÎùº

print(model.theta_[0], model.sigma_[0],'\n',model.theta_[1], model.sigma_[1],'\n',model.theta_[2], model.sigma_[2])
````

````python
# 5. Î∂ÑÎ•ò Í≤∞Í≥ºÌëú
y_pred = model.predict(X)
y_true = y
confusion_matrix(y_true, y_pred)
````

````python
# 6. Î∂ÑÎ•ò Î¶¨Ìè¨Ìä∏
print(classification_report(y_true, y_pred, target_names=iris.target_names))
````

# 2. Î≤†Î•¥ÎàÑÏù¥ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï


    - Îç∞Ïù¥ÌÑ∞ : 0 ÎòêÎäî 1Ïùò Í∞í (Ïù¥ÏßÑÎ∂ÑÎ•ò)   
        ex) ÌäπÏ†ï Îã®Ïñ¥Í∞Ä ÏûàÍ≥† ÏóÜÍ≥†Í∞Ä Îç∞Ïù¥ÌÑ∞Î°ú Ï£ºÏñ¥Ïßà Îïå (BOW : feature: ÌäπÏ†ï Îã®Ïñ¥(ÌÜ†ÌÅ∞), row: Îç∞Ïù¥ÌÑ∞ÎßàÎã§ ÏûàÍ≥†(1) ÏóÜÍ≥†(0)) 
        
    - Î≤†Î•¥ÎàÑÏù¥Î∂ÑÌè¨ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Ìòï : Î≤†Î•¥ÎàÑÏù¥ Í∞ÄÎä•ÎèÑ Î™®ÌòïÏùÑ Ï†ÅÏö©Ìïú ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®Îç∏
        Î≤†Î•¥ÎàÑÏù¥ Í∞ÄÎä•ÎèÑ Î™®Ìòï : Î≤†Î•¥ÎàÑÏù¥pdfÏóê Ï°∞Í±¥Î∂Ä ÎèÖÎ¶Ω Ï†ÅÏö©
   

**Ïã§Ïäµ : MNIST Ïà´Ïûê Î∂ÑÎ•òÎ¨∏Ï†ú**

```
(1) MNIST Ïà´Ïûê Î∂ÑÎ•òÎ¨∏Ï†úÏóêÏÑú sklearn.preprocessing.BinarizerÎ°ú xÍ∞íÏùÑ 0, 1Î°ú Î∞îÍæºÎã§
(Í∞íÏù¥ 8 Ïù¥ÏÉÅÏù¥Î©¥ 1, 8 ÎØ∏ÎßåÏù¥Î©¥ 0). Ï¶â Ìù∞ÏÉâÍ≥º Í≤ÄÏùÄÏÉâ ÌîΩÏÖÄÎ°úÎßå Íµ¨ÏÑ±Îêú Ïù¥ÎØ∏ÏßÄÎ°ú ÎßåÎì†Îã§(Îã§Ïùå ÏΩîÎìú Ï∞∏Ï°∞)

from sklearn.datasets import load_digits
digits = load_digits()
X = digits.data
y = digits.target
from sklearn.preprocessing import Binarizer
X = Binarizer(7).fit_transform(X)

Ïù¥ Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌï¥ Î≤†Î•¥ÎàÑÏù¥ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®ÌòïÏùÑ Ï†ÅÏö©ÌïòÏûê. Î∂ÑÎ•ò Í≤∞Í≥ºÎ•º Î∂ÑÎ•òÎ≥¥Í≥†ÏÑú ÌòïÏãùÏúºÎ°ú ÎÇòÌÉÄÎÇ¥Îùº.
```

````python
# 1. Ïà´Ïûê Îç∞Ïù¥ÌÑ∞ Î°úÎìú + Ïù¥ÏßÑÌôî( binarizer ) : 'Ïù¥ÏßÑÌôî'Îäî Î≤†Î•¥ÎàÑÏù¥ Î™®ÌòïÏúºÎ°ú ÏÇ¨Ïö©Ìï¥ Î∂ÑÏÑùÌïòÍ∏∞ ÏúÑÌï¥ ÏûÑÏùòÎ°ú Í∞ÄÍ≥µÌïú Í≤É

from sklearn.datasets import load_digits
digits = load_digits()
X = digits.data
y = digits.target
from sklearn.preprocessing import Binarizer
X = Binarizer(7).fit_transform(X)
````

````python
# 2. Î≤†Î•¥ÎàÑÏù¥ Î™®ÌòïÏúºÎ°ú ÌïôÏäµ

from sklearn.naive_bayes import BernoulliNB
model_bern = BernoulliNB().fit(X, y)
````

````python
# 3. Î≤†Î•¥ÎàÑÏù¥ Î™®Ìòï Î∂ÑÎ•ò Í≤∞Í≥º
# Î∂ÑÎ•òÎ≥¥Í≥†ÏÑú ÌòïÏãù

y_pred = model_bern.predict(X)
y_true = y
print(classification_report(y_true, y_pred))
````

```
Ïù¥ Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌï¥ Î≤†Î•¥ÎàÑÏù¥ ÎÇòÏù¥Î∏åÎ≤†Ïù¥Ï¶à Î™®ÌòïÏùÑ Ï†ÅÏö©ÌïòÏûê. Î∂ÑÎ•ò Í≤∞Í≥ºÎ•º Î∂ÑÎ•òÎ≥¥Í≥†ÏÑú ÌòïÏãùÏúºÎ°ú ÎÇòÌÉÄÎÇ¥Îùº.

(2) BernoulliNB ÌÅ¥ÎûòÏä§Ïùò binarize Ïù∏ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Í∞ôÏùÄ Î¨∏Ï†úÎ•º ÌíÄÏñ¥Î≥∏Îã§.
```

````python
X2 = digits.data
model_bern2 = BernoulliNB(binarize=7).fit(X2, y)
y_pred2 = model_bern2.predict(X2)
print(classification_report(y_true, y_pred2))
````
