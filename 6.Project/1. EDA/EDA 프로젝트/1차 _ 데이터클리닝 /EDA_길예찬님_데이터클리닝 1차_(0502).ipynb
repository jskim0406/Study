```python

caa={'Rating':0,'Type':'Free','Content Rating':'Everyone','Current Ver':'Varies with device','Android Ver':'Varies with device'}
ad=a1.fillna(value=caa)
ad
```

```python
len(ad[ad["Rating"].isnull()])
```

```python
msno.bar(ad)
```

```python
ad
```

```python
ad.drop(['Last Updated','Current Ver','Android Ver'],axis=1,inplace=True)
```

```python
ad
```

```python
ad.columns = ad.columns.str.replace(" ","_")
```

```python
ad['Content_Rating'].unique()
```

```python


ad.Size = ad.Size.replace("Varies with device",np.nan)
ad.Size = ad.Size.str.replace("M","000")
ad.Size = ad.Size.str.replace("k","")
#store.Size = store.Size.apply(lambda x: float(x.replace("k",""))/1000 if "k" in x else x)
#ı wanted use this but ı get error so ı added "000" to Megabyte

ad.Size = ad.Size.replace("1,000+",1000)

ad.Installs = ad.Installs.str.replace(",","")
ad.Installs = ad.Installs.apply(lambda x: x.strip("+"))
ad.Installs = ad.Installs.replace("Free",np.nan)

ad.Reviews = ad.Reviews.replace("Varies with device",np.nan)
ad.Reviews = ad.Reviews.str.replace("M","000000")

ad.Price = ad.Price.str.replace("$","")

ad = ad.drop(ad.index[10472])



ad[["Size","Installs","Reviews","Price"]] = ad[["Size","Installs","Reviews","Price"]].astype("float")
```

```python
ad.corr()
```

```python
f,ax = plt.subplots(figsize = (10,10))
sns.heatmap(ad.corr(), annot = True, linewidths = .7, fmt = ".2f", ax=ax)
```
